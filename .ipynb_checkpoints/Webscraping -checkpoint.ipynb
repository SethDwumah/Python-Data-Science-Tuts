### Cleaning and merging scraped data with Pandas

```python
#shooting.head()
```

```python
drop upper row
shooting.columns =shooting.columns.droplevel()
```

```python
shooting.head()
```

```python
matches[0].head()
```

```python
# merge the two tables

chelsea_data = matches[0].merge(shooting[["Date","Sh","SoT","Dist","FK","PK","PKatt"]], on="Date")
```

```python
chelsea_data.head()
```

```python
# drop Notes
chelsea_data.drop(['Notes'],axis=1, inplace=True)
```
