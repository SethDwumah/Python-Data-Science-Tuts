### Treating missing values

#### Figuring out what data is missing

```python
missing = np.nan

series_obj = Series(['row 1','row 2',missing,'row 4','row 5',missing,'row 6','row 7',missing])
series_obj
```

```python
series_obj.isnull()
```

#### Filling in for missing values

```python
np.random.seed(25)
DF = DataFrame(np.random.rand(36).reshape(6,6))
DF
```

```python
DF.loc[3:5, 0]= missing
DF.loc[1:4, 5] = missing
DF
```

```python
DF.loc[2:3, 3]= missing
DF
```

```python
filled_DF = DF.fillna(0)
filled_DF
```

```python
filled_DF = DF.fillna({0: 0.1, 5: 1.25})
filled_DF
```

```python
filled_DF = DF.fillna({0:0.1, 3: 2.2, 5:1.24})
filled_DF
```

```python
filled_DF = DF.fillna(method = 'ffill')
filled_DF
```

#### Counting missing values

```python
np.random.seed(25)
DF = DataFrame(np.random.rand(36).reshape(6,6))

DF.loc[3:5, 0]= missing
DF.loc[1:4, 5] = missing
DF
```

```python
DF.isnull().sum()
```

#### Filtering out missing values

```python
DF_no_NaN = DF.dropna()
DF_no_NaN
```

```python
DF_no_NaN = DF.dropna(axis=1)
DF_no_NaN
```
