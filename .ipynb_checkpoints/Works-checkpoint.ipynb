# Basic SQL Queries

```python
import pandas as pd
```

```python
titanic_dataset=pd.read_csv("C:\\Users\\Administrator\\Desktop\\data\\SQL-for-Data-Analytics-main\\titanic.csv")
print(titanic_dataset.shape, titanic_dataset[['PassengerId']].drop_duplicates().shape)
titanic_dataset.head()
```

```python
#Selecting everything from dataset

ps.sqldf("""

SELECT
    *
FROM titanic_dataset



""")
```

```python
# Giving as alias name to the dataset and using it in "Select" statement
ps.sqldf("""

Select
   df. *
FROM titanic_dataset AS df    



""")
```

```python
#Using "Limit " to get top 5 rows
ps.sqldf("""
Select
    df.*
    From titanic_dataset AS df
    Limit 5

""")
```

```python
# Taking only Passenger Id and the name from the dataset
ps.sqldf("""
Select
    df.PassengerId,
    df.Name
From titanic_dataset AS df
Limit 6



""")
```

```python
# Using Alias to name the columns
ps.sqldf("""
Select 
    df.PassengerId AS ID_of_Passenger,
    df.Name AS Name_of_Passenger
FROM titanic_dataset AS df
Limit 6


""")
```

## Where

```python
# Selecting everything from the dataset for passengers
ps.sqldf("""
Select
df.*
From titanic_dataset AS df
Where df.age < 1


""")
```

```python
# Select "None " in the Cabin

ps.sqldf("""

Select 
    df.*
    
From titanic_dataset AS df
where df.age IS NULL and df.Cabin Is Null


""")
```

```python
# Selecting from the dataset using between
ps.sqldf("""
Select 
    df.Name as Name_of_Passenger,
    df.age as Age_of_Passenger, 
    df.Sex as Sex_of_Passenger
 From titanic_dataset as df
 where df.age >1 
     AND df.Sex = "female"

""")
```

```python
# Selecting everything from the dataset for passengers 
#(age is above 70 years and Pclass is 1)

ps.sqldf("""
Select
     df.*
From titanic_dataset AS df
Where (df.age<1 And df.Sex="female")
OR (df.age >70 And df.Pclass =1 )


""")
```

```python
# Selecting everything from the dataset for the passengets
ps.sqldf("""

Select
    df.*
From titanic_dataset AS df
Where df.Cabin In ('A5', 'A6','A7')



""")
```

```python
ps.sqldf("""

Select
    df.*
From titanic_dataset as df
where df.Age In (65,70,71)



""")
```

### Using Like eg. LIKE "%ABC%"

```python
#Selecting from the dataset
ps.sqldf("""
Select
    df.Name,
    df.Age
From titanic_dataset as df
where df.Name LIKE "%MRS.%"
Limit 5



""")
```

```python
#Selecting from the dataset
ps.sqldf("""
Select
    df.Name,
    df.Age
From titanic_dataset as df
where df.Name LIKE "%MR.%"
Limit 5



""")
```
