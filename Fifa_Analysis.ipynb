```python
# number of players in each country
fifa['nationality'].value_counts()
```

```python
player_country=fifa['nationality'].value_counts()[:10]
player_country
```

```python
player_country.keys()
```

```python
#fifa['nationality'].value_counts()[:10].keys()
```

```python
#top 10 countries that has more players in fifa
plt.figure(figsize=(8,5))
plt.bar(list(player_country.keys()),list(player_country),color='g')
plt.xlabel('Nationality', fontsize=15)
plt.ylabel('Number of Players', fontsize=15)
plt.title('Top 10 Countries With Most Players in Fifa',fontsize=20, color='r')
plt.show
        
```

```python
# player salary
player_salary = fifa[['short_name','wage_eur']]
player_salary.head()
```

```python
player_salary = player_salary.sort_values(by=['wage_eur'], ascending= False)[:5]
player_salary.head()
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(player_salary['short_name']),list(player_salary['wage_eur']), color= ['g','b','r','pink','orange'])
plt.xlabel('Players',fontsize=15 )
plt.ylabel('Wage', fontsize=15)
plt.title('Top 5 Players that Receive Huge Wage', fontsize=20)
plt.show()
```
