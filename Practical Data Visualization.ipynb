## Defining elements of a plot


#### Defining axes, ticks, and grids

```python
x1 = range(1,10)
y1 = [0,2,3,4,1,4,3,2,1]
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.plot(x1,y1)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.set_xticks([0,1,2,4,5,6,8,9,10])
ax.set_yticks([0,1,2,3,4,5])

ax.plot(x1,y1)
```

```python
fig = plt.figure()

ax =fig.add_axes([.1,.1,1,1])
ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.grid()
ax.plot(x1,y1)
```

#### Generating multiple plots in one figure with subplots

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1,2)

ax1.plot(x1)
ax2.plot(x1,y1)
```
